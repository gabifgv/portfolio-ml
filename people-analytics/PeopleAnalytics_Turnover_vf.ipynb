## 5. Preparacao dos Dados e SMOTE

```python
# Codificar variavel-alvo
df['Turnover_bin'] = (df['Turnover'] == 'turnover').astype(int)

# Codificar variaveis categoricas com LabelEncoder
cat_cols = df.select_dtypes(include=['object','string']).columns.drop('Turnover')
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# Separar features e target
X = df.drop(columns=['Turnover', 'Turnover_bin', 'Turnover_bin_eda'], errors='ignore')
y = df['Turnover_bin']

# Split 70/30 estratificado
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42, stratify=y)

# SMOTE apenas no treino (treino final)
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print(f"Treino original: {len(y_train)} | Teste: {len(y_test)}")
print(f"Antes SMOTE: {dict(y_train.value_counts())}")
print(f"Apos SMOTE:  {dict(y_train_res.value_counts())}")
```

**Interpretacao:** O SMOTE equilibrou as classes no treino, criando amostras sinteticas de turnover. O teste permanece inalterado para avaliacao realista.

## 6. Funcao Auxiliar: Otimizacao de Threshold via Cross-Validation

```python
# ============================================================
# CORRECAO METODOLOGICA: threshold otimizado no TREINO via CV
# evita data leakage (buscar threshold no conjunto de teste)
# ============================================================

def threshold_cv(estimator, X_tr, y_tr, n_splits=5):
    """
    Encontra o threshold otimo via validacao cruzada estratificada no treino.
    SMOTE aplicado dentro de cada fold para evitar data leakage.
    """
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    thresholds = np.arange(0.10, 0.91, 0.01)
    f1_por_threshold = np.zeros(len(thresholds))

    for fold_train_idx, fold_val_idx in skf.split(X_tr, y_tr):
        X_fold_train = X_tr.iloc[fold_train_idx]
        y_fold_train = y_tr.iloc[fold_train_idx]
        X_fold_val   = X_tr.iloc[fold_val_idx]
        y_fold_val   = y_tr.iloc[fold_val_idx]

        smote_cv = SMOTE(random_state=42)
        X_res, y_res = smote_cv.fit_resample(X_fold_train, y_fold_train)

        estimator.fit(X_res, y_res)
        prob_val = estimator.predict_proba(X_fold_val)[:, 1]

        for i, t in enumerate(thresholds):
            pred = (prob_val >= t).astype(int)
            if pred.sum() == 0:
                continue
            f1_por_threshold[i] += f1_score(y_fold_val, pred, zero_division=0)

    f1_medio = f1_por_threshold / n_splits
    best_idx = np.argmax(f1_medio)
    best_t   = round(thresholds[best_idx], 2)
    print(f"  Threshold otimo (CV 5-fold): {best_t} | F1 medio CV: {f1_medio[best_idx]:.4f}")
    return best_t


def plot_cm(cm, titulo, cor='Greys'):
    fig, ax = plt.subplots(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap=cor,
                xticklabels=['Permanece', 'Saiu'],
                yticklabels=['Permanece', 'Saiu'], ax=ax)
    ax.set_xlabel('Predito')
    ax.set_ylabel('Real')
    ax.set_title(titulo)
    plt.tight_layout()
    plt.show()
```

## 7. Modelo 1: Regressao Logistica

```python
print("Buscando threshold otimo via CV...")
t_log = threshold_cv(LogisticRegression(max_iter=1000, random_state=42), X_train, y_train)

log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(X_train_res, y_train_res)
prob_log = log_model.predict_proba(X_test)[:, 1]
pred_log = (prob_log >= t_log).astype(int)
cm_log = confusion_matrix(y_test, pred_log)

print(f"Threshold (CV): {t_log}")
print(f"Precisao: {precision_score(y_test, pred_log, zero_division=0):.4f} | Recall: {recall_score(y_test, pred_log, zero_division=0):.4f} | F1: {f1_score(y_test, pred_log, zero_division=0):.4f}")
print(f"AUC-ROC: {roc_auc_score(y_test, prob_log):.4f}")
print()
print(classification_report(y_test, pred_log, target_names=['Permanece', 'Saiu']))
plot_cm(cm_log, f'Regressao Logistica (threshold={t_log})')
```

---
## 8. Modelo 2: Random Forest

```python
print("Buscando threshold otimo via CV...")
t_rf = threshold_cv(RandomForestClassifier(n_estimators=300, max_depth=10, random_state=42, n_jobs=-1), X_train, y_train)

rf_model = RandomForestClassifier(n_estimators=300, max_depth=10, random_state=42, n_jobs=-1)
rf_model.fit(X_train_res, y_train_res)
prob_rf = rf_model.predict_proba(X_test)[:, 1]
pred_rf = (prob_rf >= t_rf).astype(int)
cm_rf = confusion_matrix(y_test, pred_rf)

print(f"Threshold (CV): {t_rf}")
print(f"Precisao: {precision_score(y_test, pred_rf, zero_division=0):.4f} | Recall: {recall_score(y_test, pred_rf, zero_division=0):.4f} | F1: {f1_score(y_test, pred_rf, zero_division=0):.4f}")
print(f"AUC-ROC: {roc_auc_score(y_test, prob_rf):.4f}")
print()
print(classification_report(y_test, pred_rf, target_names=['Permanece', 'Saiu']))
plot_cm(cm_rf, f'Random Forest (threshold={t_rf})')
```

## 10. Modelo 3: XGBoost

```python
print("Buscando threshold otimo via CV...")
t_xgb = threshold_cv(xgb.XGBClassifier(n_estimators=300, max_depth=6, learning_rate=0.1,
                                         random_state=42, eval_metric='logloss'), X_train, y_train)

xgb_model = xgb.XGBClassifier(n_estimators=300, max_depth=6, learning_rate=0.1,
                                random_state=42, eval_metric='logloss')
xgb_model.fit(X_train_res, y_train_res)
prob_xgb = xgb_model.predict_proba(X_test)[:, 1]
pred_xgb = (prob_xgb >= t_xgb).astype(int)
cm_xgb = confusion_matrix(y_test, pred_xgb)

print(f"Threshold (CV): {t_xgb}")
print(f"Precisao: {precision_score(y_test, pred_xgb, zero_division=0):.4f} | Recall: {recall_score(y_test, pred_xgb, zero_division=0):.4f} | F1: {f1_score(y_test, pred_xgb, zero_division=0):.4f}")
print(f"AUC-ROC: {roc_auc_score(y_test, prob_xgb):.4f}")
print()
print(classification_report(y_test, pred_xgb, target_names=['Permanece', 'Saiu']))
plot_cm(cm_xgb, f'XGBoost (threshold={t_xgb})')
```

## 11. Modelo 4: Rede Neural (MLP)

```python
print("Buscando threshold otimo via CV...")
t_nn = threshold_cv(MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500,
                                   random_state=42, early_stopping=True), X_train, y_train)

nn_model = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500,
                         random_state=42, early_stopping=True)
nn_model.fit(X_train_res, y_train_res)
prob_nn = nn_model.predict_proba(X_test)[:, 1]
pred_nn = (prob_nn >= t_nn).astype(int)
cm_nn = confusion_matrix(y_test, pred_nn)

print(f"Threshold (CV): {t_nn}")
print(f"Precisao: {precision_score(y_test, pred_nn, zero_division=0):.4f} | Recall: {recall_score(y_test, pred_nn, zero_division=0):.4f} | F1: {f1_score(y_test, pred_nn, zero_division=0):.4f}")
print(f"AUC-ROC: {roc_auc_score(y_test, prob_nn):.4f}")
print()
print(classification_report(y_test, pred_nn, target_names=['Permanece', 'Saiu']))
plot_cm(cm_nn, f'Rede Neural MLP (threshold={t_nn})')
```

## 12. Comparacao Final dos Modelos

```python
def metricas(prob, real, threshold, nome):
    pred = (prob >= threshold).astype(int)
    return {
        'Modelo': nome, 'Threshold': threshold,
        'Precisao': round(precision_score(real, pred, zero_division=0), 4),
        'Recall': round(recall_score(real, pred, zero_division=0), 4),
        'F1-Score': round(f1_score(real, pred, zero_division=0), 4),
        'Acuracia Bal.': round(balanced_accuracy_score(real, pred), 4),
        'AUC-ROC': round(roc_auc_score(real, prob), 4)
    }

comp = pd.DataFrame([
    metricas(prob_log, y_test, t_log, 'Regressao Logistica'),
    metricas(prob_rf,  y_test, t_rf,  'Random Forest'),
    metricas(prob_xgb, y_test, t_xgb, 'XGBoost'),
    metricas(prob_nn,  y_test, t_nn,  'Rede Neural (MLP)')
])

print("=" * 90)
print("COMPARACAO FINAL DOS MODELOS")
print("=" * 90)
print(comp.to_string(index=False))

vencedor = comp.loc[comp['F1-Score'].idxmax(), 'Modelo']
print(f"\n>>> Modelo recomendado (maior F1-Score): {vencedor}")
```

```python
from IPython.display import Markdown, display
display(Markdown("""> **Nota metodologica:** Os thresholds foram otimizados via validacao cruzada
> estratificada (5-fold) no conjunto de treino, com SMOTE aplicado dentro de
> cada fold. O conjunto de teste foi utilizado exclusivamente para avaliacao
> final, sem nenhuma informacao vazar para a selecao de hiperparametros.
> Essa abordagem evita data leakage e produz estimativas de desempenho
> mais realistas para producao."""))
```

```python
# Grafico comparativo de metricas
metricas_cols = ['Precisao', 'Recall', 'F1-Score', 'Acuracia Bal.', 'AUC-ROC']
x = np.arange(len(metricas_cols))
width = 0.2
colors = ['#333333', '#666666', '#999999', '#BBBBBB']

fig, ax = plt.subplots(figsize=(14, 5))
for i, row in comp.iterrows():
    bars = ax.bar(x + i*width, row[metricas_cols].values, width, label=row['Modelo'], color=colors[i])
    for bar in bars:
        ax.annotate(f'{bar.get_height():.2f}', xy=(bar.get_x()+bar.get_width()/2, bar.get_height()),
                    ha='center', va='bottom', fontsize=8)

ax.set_xticks(x + width*1.5)
ax.set_xticklabels(metricas_cols)
ax.set_ylim(0, 1.1)
ax.set_title('Comparacao de Metricas - 4 Modelos')
ax.legend()
plt.tight_layout()
plt.show()
```
